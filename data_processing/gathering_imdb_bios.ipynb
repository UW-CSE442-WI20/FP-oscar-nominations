# Gathering IMDB Bios for converted JSONs

## Importing Packages

```python
import pandas as pd
import pathlib as path
from googlesearch import search
import time
```

## Importing Files

```python
rel_path = path.Path().absolute()
oscars = pd.read_csv(str(rel_path) + r"\raw_data\all_nominees.csv")
ids = oscars[['Name', 'Movie', 'Year of award']]
```

## Searching Google for Bio IMDB Codes

```python
tic = time.perf_counter()
ids['bioID'] = ''
ids['year_awarded'] = ids['Year of award']
for i, row in ids.iterrows():
    print(f'{row.Movie} {row.year_awarded - 1} imdb')
    url = search(f'{row.Name} imdb', tld='com', lang='en', num=1, start=0, stop=1, pause=2.0)
    for link in url:
        ids.set_value(i,'bioID',link)
toc = time.perf_counter()
print(f'Took {toc - tic:0.4f} Seconds')
```

## Searching Google for Title IMDB Codes

```python
tic = time.perf_counter()
ids['movieID'] = ''
ids['year_awarded'] = ids['Year of award']
for i, row in ids.iterrows():
    print(f'{row.Movie} {row.year_awarded - 1} imdb')
    url = search(f'{row.Movie} {row.year_awarded - 1} imdb', tld='com', lang='en', num=1, start=0, stop=1, pause=2.0)
    for link in url:
        ids.set_value(i,'movieID',link)
toc = time.perf_counter()
print(f'Took {toc - tic:0.4f} Seconds')
```
